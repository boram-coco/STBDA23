# 2. Imports

```python
import numpy as np
import matplotlib.pyplot as plt
import sklearn.tree
#---#
import warnings
warnings.filterwarnings('ignore')
```

# 3. `weights`

`# 예제1` – 아래의 예제를 생각하자.

```python
X = np.array([1,2,3,4,5,6,7]).reshape(-1,1)
y = np.array([10,11,12,20,21,22,23])
```

```python
plt.plot(X,y,'o')
```

`-` 어디에서 분기점을 나누는게 좋을까? 당연히 3.5 정도..

```python
predictr = sklearn.tree.DecisionTreeRegressor(max_depth=1)
predictr.fit(X,y)
sklearn.tree.plot_tree(predictr);
```

`# 예제2` – 아래의 예제를 생각하자.

```python
X = np.array([1]*5000+[2]*5000+[3,4,5,6,7]).reshape(-1,1)
y = np.array([10]*5000+[11]*5000+[12,20,21,22,23])
```

```python
plt.plot(X,y,'o',alpha=0.1)
```

`-` 분기점은 어디에 나누는게 좋을까? 이 경우는 1.5근처에서 나누는게
합리적으로 보임

```python
predictr = sklearn.tree.DecisionTreeRegressor(max_depth=1)
```

```python
predictr.fit(X,y)
```

```python
sklearn.tree.plot_tree(
    predictr
)
```

`# 예제3` – 다시 아래의 예제를 생각하자.

```python
X = np.array([1,2,3,4,5,6,7]).reshape(-1,1)
y = np.array([10,11,12,20,21,22,23])
```

```python
plt.plot(X,y,'o')
plt.plot(X[:2],y[:2],'o')
```

-   주황색점들을 잘 맞추는 것이 파란색점들을 잘 맞추는 것보다 5000배정도
    중요하다고 상상하자.

```python
predictr = sklearn.tree.DecisionTreeRegressor(max_depth=1)
predictr.fit(X,y,sample_weight=[5000,5000,1,1,1,1,1])
sklearn.tree.plot_tree(predictr);
```
